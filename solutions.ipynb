**Q5** _Use the coins data in the sample dataset and apply a Hough transform to detect circles._

Let us peak into the edge detection part first

```python
from skimage import data
from skimage.filter import canny
from skimage import img_as_ubyte

coins = img_as_ubyte(data.coins()[0:95, 70:370])
edges = canny(coins, sigma=3, low_threshold=10, high_threshold=50)
plt.imshow(edges, cmap=plt.cm.gray)
```

ok, now let us find the accumulator bins. We will compute these bins for radius ranging from 15 to 30 at steps of 2. Let us take a look at the bin for r=30

```python
from skimage.transform import hough_circle

hough_radii = np.arange(15, 30, 2)
hough_res = hough_circle(edges, hough_radii)

plt.imshow(hough_res[-1, :], cmap=plt.cm.gray)
```

That works! Clearly, the peak seems to coincide with the center of the last circle. But how can we get the coordinates of the peak automatically?

```python
from skimage.feature import peak_local_max

center = peak_local_max(hough_res[-1, ...], num_peaks=1)
```

Now, we can plot a circle on the image

```python
plt.imshow(coins, plt.cm.gray)

cx, cy = center[0][1], center[0][0]
plt.plot(cx, cy, 'ro');  # plot center
plt.plot(cx, cy, 'go', markersize=hough_radii[-1]*2, markerfacecolor='none',  # plot circle
         markeredgecolor='g', markeredgewidth=3)
plt.axis('off')
```

To find the other circles, you should collect the peaks for all the radii (15 in total) and plot the 5 most prominent peaks.

```python
# collecting all the peaks and the peak values
center_x, center_y, peak_values = [], [], []
for h in hough_res:
    center = peak_local_max(h, num_peaks=1)[0]
    center_y.append(center[0])
    center_x.append(center[1])

    peak_values.append(h[center[0], center[1]])

plt.imshow(coins, plt.cm.gray)    

# find the max
for idx in np.argsort(peak_values)[::-1][:5]:
    plt.plot(center_x[idx], center_y[idx], 'ro')
    plt.plot(center_x[idx], center_y[idx], 'go', markersize=hough_radii[idx]*2,
             markerfacecolor='none', markeredgecolor='g', markeredgewidth=3)
plt.axis('off')
```

**Q6** _Can you plot the cross validation accuracies as a function of the parameter C (1, 10, 100, 1000) for a linear kernel?_

```python
from sklearn.svm import SVC
from sklearn.cross_validation import cross_val_score, StratifiedKFold
from sklearn import datasets

digits = datasets.load_digits()
cv = StratifiedKFold(digits.target, 5)  # we can also use ShuffleSplit or any of the other iterators
C = np.logspace(-8, 3, 12)

scores = []
for this_C in C:
    print('C = 1e%d' % np.log10(this_C))
    clf = SVC(kernel='linear', C=this_C)
    scores.append(cross_val_score(clf, X=digits.data, y=digits.target, cv=cv).mean())

import matplotlib.pyplot as plt
plt.semilogx(C, scores, 'bo-', markeredgewidth=2, markerfacecolor='w', markeredgecolor='b', markersize=8,
             linewidth=2.)
plt.ylabel('Mean accuracy', fontsize=14)
plt.xlabel('C', fontsize=14)
```
